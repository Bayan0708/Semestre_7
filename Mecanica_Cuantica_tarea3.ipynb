```python
import matplotlib.pyplot as plt
import numpy as np

# Constantes dadas en el problema
eB_2mec = 1  # Representación simbólica de eB/(2*m_e*c) para simplificar la gráfica
hbar = 1  # Valor de h-bar en unidades naturales para simplificar

# Valores de m_j para cada caso
mj_3_2 = np.array([-3/2, -1/2, 1/2, 3/2])
mj_5_2 = np.array([-5/2, -3/2, -1/2, 1/2, 3/2, 5/2])

# Energías para j=3/2 y j=5/2
E_j_3_2 = (4/5) * eB_2mec * hbar * mj_3_2
E_j_5_2 = (6/5) * eB_2mec * hbar * mj_5_2

# Crear la gráfica
plt.figure(figsize=(10, 6))
plt.plot(mj_3_2, E_j_3_2, 'o', color="blue", label=r'$j = 3/2$')
plt.plot(mj_5_2, E_j_5_2, 'o', color="red", label=r'$j = 5/2$')

# Etiquetas y títulos
plt.xlabel(r'$m_j$')
plt.ylabel(r'$E_n^{(1)}$ (unidades de $\frac{eB}{2m_e c} \hbar$)')
plt.title('Niveles de Energía con el Efecto Zeeman para $l=2$')
plt.axhline(0, color='black', linewidth=0.5)
plt.legend()

plt.grid(True)
plt.show()
```
