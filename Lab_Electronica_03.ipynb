```python
import numpy as np
import matplotlib.pyplot as plt

# Valores de x en milivoltios y conversión a voltios
x_values_mv = np.array([0, 180, 500, 800, 1360, 1720, 340, 680, 980, 1440])  # en mV
x_values_v = x_values_mv / 1000  # conversión a voltios

# Constantes
denominator = 240 / 1000  # 240 mV convertido a V
resistance_factor = 1  # kΩ

# Cálculo de R_f
R_f = (x_values_v / denominator) * resistance_factor  # en kΩ

# Gráfica
plt.figure(figsize=(8, 5))
plt.plot(x_values_mv, R_f, marker='o', linestyle='-', color='b', label='R_f vs x')
plt.title('Relación entre $R_f$ y $x$', fontsize=14)
plt.xlabel('x (mV)', fontsize=12)
plt.ylabel('$R_f$ (kΩ)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(fontsize=12)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Valores de x en milivoltios (mV)
x_values_mv = np.array([0, 180, 500, 800, 1360, 1720, 340, 680, 980, 1440])  # en mV
x_values_v = x_values_mv / 1000  # Convertir a voltios

# Constante de denominador (240 mV = 0.240 V)
denominator = 240 / 1000  # 240 mV en voltios

# Cálculo de R_f
R_f = (x_values_v / denominator)  # en kΩ

# Realizar ajuste lineal
slope, intercept, r_value, p_value, std_err = linregress(x_values_mv, R_f)

# Crear la gráfica
plt.figure(figsize=(8, 5))
plt.plot(x_values_mv, R_f, marker='o', linestyle='-', color='b', label='Datos de $R_f$')
plt.plot(x_values_mv, slope * x_values_mv + intercept, color='r', label=f'Ajuste lineal: $R_f = {slope:.3f} x + {intercept:.3f}$')

# Etiquetas y título
plt.title('Ajuste lineal de $R_f$ vs $x$', fontsize=14)
plt.xlabel('x (mV)', fontsize=12)
plt.ylabel('$R_f$ (kΩ)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(fontsize=12)
plt.tight_layout()

# Mostrar la gráfica
plt.show()

# Imprimir la ecuación del ajuste lineal
print(f'Ecuación de ajuste lineal: R_f = {slope:.3f} * x + {intercept:.3f}')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Datos extraídos de la tabla
rf = np.array([0.18, 0.34, 0.50, 0.68, 0.80, 0.98, 1.36, 1.44, 1.72])  # Voltaje pico a pico (V)
vout = np.array([0.64, 1.35, 2.28, 2.86, 3.48, 4.00, 5.61, 5.81, 7.01])    # Resistencia variable (kΩ)

# Ajuste lineal
slope, intercept, r_value, p_value, std_err = linregress(vout, rf)
rf_fit = slope * vout + intercept

# Graficamos los datos y el ajuste lineal
plt.figure(figsize=(8, 6))
plt.scatter(vout, rf, color='blue', label='Datos experimentales')
plt.plot(vout, rf_fit, color='red', linestyle='--', label=f'Ajuste lineal: Rf = {slope:.2f}Vout + {intercept:.2f}')
plt.title('Relación entre Vout (V) y Rf (kΩ)', fontsize=14)
plt.xlabel('Vout (V pico a pico)', fontsize=12)
plt.ylabel('Rf (kΩ)', fontsize=12)
plt.legend(fontsize=10)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

# Mostramos los parámetros del ajuste
slope, intercept, r_value**2
```
