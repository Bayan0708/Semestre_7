```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos las constantes
g = 9.81  # aceleración de la gravedad (m/s^2)
m = 0.1  # masa del imán (kg)
k = 0.5  # constante de frenado magnético (kg/s)

# Definimos el tiempo
t = np.linspace(0, 10, 500)  # tiempo en segundos

# Solución de la ecuación diferencial: velocidad en función del tiempo
v_terminal = (m * g) / k  # velocidad terminal
v = v_terminal * (1 - np.exp(-k * t / m))  # velocidad del imán

# Graficamos la velocidad en función del tiempo
plt.figure(figsize=(8, 6))
plt.plot(t, v, label='Velocidad del imán', color='b', lw=2)
plt.axhline(y=v_terminal, color='r', linestyle='--', label='Velocidad terminal')
plt.title('Movimiento desacelerado del imán en el tubo de cobre', fontsize=14)
plt.xlabel('Tiempo (s)', fontsize=12)
plt.ylabel('Velocidad (m/s)', fontsize=12)
plt.legend()
plt.grid(True)
plt.show()
```
