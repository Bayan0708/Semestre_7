```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Datos de la tabla
data = {
    "Pulsera Inteligente": [85, 82, 83, 81, 80, 85, 84, 83, 83, 83],
    "Proyecto": [88, 83, 85, 85, 86, 95, 90, 85, 84, 87]
}

# Crear un DataFrame
df = pd.DataFrame(data)

# C치lculo de promedios
prom_pulsera = df["Pulsera Inteligente"].mean()
prom_proyecto = df["Proyecto"].mean()

# C치lculo de desviaci칩n est치ndar
std_pulsera = df["Pulsera Inteligente"].std()
std_proyecto = df["Proyecto"].std()

# Mostrar resultados
print(f"Promedio Pulsera Inteligente: {prom_pulsera:.2f}, Desviaci칩n est치ndar: {std_pulsera:.2f}")
print(f"Promedio Proyecto: {prom_proyecto:.2f}, Desviaci칩n est치ndar: {std_proyecto:.2f}")

# Visualizaci칩n
plt.figure(figsize=(10, 6))
plt.plot(df["Pulsera Inteligente"], label="Pulsera Inteligente", marker="o")
plt.plot(df["Proyecto"], label="Proyecto", marker="o")
plt.axhline(prom_pulsera, color="blue", linestyle="--", label="Promedio Pulsera")
plt.axhline(prom_proyecto, color="orange", linestyle="--", label="Promedio Proyecto")
plt.title("Comparaci칩n del Pulso Cardiaco [rpm]")
plt.xlabel("Medici칩n")
plt.ylabel("Pulso [rpm]")
plt.legend()
plt.grid(True)
plt.show()
```
