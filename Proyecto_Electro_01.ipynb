```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
mu_0 = 4 * np.pi * 1e-7  # Permeabilidad del vacío (T·m/A)
I = 10  # Corriente en la bobina (A)
R = 0.05  # Radio de la bobina (m)
n_turns = 100  # Número de vueltas de la bobina

# Función para calcular el campo magnético en un punto a distancia z
def magnetic_field(z, R, I, n_turns):
    B = (mu_0 * I * n_turns * R**2) / (2 * (R**2 + z**2)**(3/2))
    return B

# Valores de distancia desde la bobina
z = np.linspace(0.01, 1, 500)  # Distancia de 1 cm a 1 metro

# Campo magnético en función de la distancia
B = magnetic_field(z, R, I, n_turns)

# Gráfico
plt.figure(figsize=(8,6))
plt.plot(z, B, label=f'I = {I} A, R = {R} m, n_turns = {n_turns}')
plt.title('Campo magnético generado por una bobina')
plt.xlabel('Distancia desde la bobina (m)')
plt.ylabel('Campo magnético (T)')
plt.grid(True)
plt.legend()
plt.show()
```
