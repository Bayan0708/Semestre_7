```python
import matplotlib.pyplot as plt
import numpy as np

# Par치metros del circuito
I_0 = 1.89  # Corriente inicial en amperios (estado estacionario)
tau = 8 / (40 + 48)  # Constante de tiempo en segundos

# Tiempo desde t = -0.1 s hasta t = 1 s
t_vals = np.linspace(-0.1, 1, 500)

# Funci칩n de corriente i(t) para t < 0 y decay para t >= 0
i_vals = [I_0 if t < 0 else I_0 * np.exp(-t / tau) for t in t_vals]

# Graficar la corriente en funci칩n del tiempo
plt.figure(figsize=(8, 6))
plt.plot(t_vals, i_vals, label=r'$i_1(t)$', color='b')
plt.axvline(x=0, color='r', linestyle='--', label='t = 0 s')
plt.title(r'Corriente $i_1$ en funci칩n del tiempo')
plt.xlabel('Tiempo (s)')
plt.ylabel('Corriente $i_1$ (A)')
plt.grid(True)
plt.legend()
plt.xlim(-0.1, 1)
plt.ylim(0, 2)
plt.show()
```
