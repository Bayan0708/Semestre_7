```python
import numpy as np
import matplotlib.pyplot as plt
# Definimos el rango de valores de x e y
x = np.linspace(-5, 5, 400)
y = np.linspace(-5, 5, 400)
X, Y = np.meshgrid(x, y)
# Definimos la constante A
A = 1
# Calculamos la función de corriente psi = 2 * A * X * Y
psi = 2 * A * X * Y
# Creamos el gráfico de líneas de corriente
plt.figure(figsize=(8, 8))
plt.contour(X, Y, psi, levels=np.linspace(-25, 25, 50), cmap="coolwarm")
# Añadimos etiquetas y detalles del gráfico
plt.xlabel("x")
plt.ylabel("y")
plt.title("Líneas de corriente del flujo con $F(z) = A z^2$")
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.colorbar(label='$\psi(x, y) = 2 A x y$')
plt.show()
```

```python
# Recalcular el campo de velocidades
U_corrected = 2 * A * X
V_corrected = -2 * A * Y
# Creación del gráfico actualizado
plt.figure(figsize=(8, 8))
contour = plt.contour(X, Y, psi, levels=10, colors="blue", linestyles="--")
plt.clabel(contour, inline=1, fontsize=10, fmt="%.1f")
plt.streamplot(X, Y, U_corrected, V_corrected, color="red", linewidth=1, density=1.2, arrowstyle='->')
# Configuración del gráfico
plt.title("Líneas de corriente del flujo con $F(z) = A z^2$")
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
